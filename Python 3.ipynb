```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

df = pd.read_csv("conductivity data_studygroup3.csv")   # reads file into a data frame effectively a csv in python
df.head() # displays first 5 rows of data frame
```

```python
# Scatter plots
plt.figure (figsize=(20,10))

plt.subplot(231)
plt.scatter(df['30 um Measured Grain Size'], df['30 um Measured Conductivity'], s = 100,  color = 'red', marker = 'o', ec = 'black') # instead of addressing an array by specifying the column number we can instead call it by name
plt.title('Process 1')
plt.ylabel('Conductivity (S/m)')
plt.xlabel('Grain Size (um)')

plt.subplot(232)
plt.scatter(df['50 um Measured Grain Size'], df['50 um Measured Conductivity'], s = 100,  color = 'green', marker = '^', ec = 'black')
plt.title('Process 2')
plt.ylabel('Conductivity (S/m)')
plt.xlabel('Grain Size (um)')

plt.subplot(233)
plt.scatter(df['100 um Measured Grain Size'], df['100 um Measured Conductivity'], s = 100,  color = 'blue', marker = 'd', ec = 'black')
plt.title('Process 3')
plt.ylabel('Conductivity (S/m)')
plt.xlabel('Grain Size (um)')

plt.subplot(234)
plt.scatter(df['150 um Measured Grain Size'], df['150 um Measured Conductivity'], s = 100,  color = 'yellow', marker = 'p', ec = 'black')
plt.title('Process 4')
plt.ylabel('Conductivity (S/m)')
plt.xlabel('Grain Size (um)')

plt.subplot(235)
plt.scatter(df['200 um Measured Grain Size'], df['200 um Measured Conductivity'], s = 100,  color = 'purple', marker = 's', ec = 'black')
plt.title('Process 5')
plt.ylabel('Conductivity (S/m)')
plt.xlabel('Grain Size (um)')
```

```python
df_means = df.mean()    # takes means of all columns of data 
df_std = df.std()   # takes means of all columns of data 

print(df_means)
print(df_std)

plt.figure (figsize=(15,10))
plt.errorbar(df_means[0:10:2], df_means[1:10:2], yerr = df_std[1:10:2], xerr = df_std[0:10:2], fmt = 'ro-', linewidth = 2, capsize = 6) # iterates through the rows of the means array to separate out grain size and conductivity
plt.xlabel('Grain Size (um)')
plt.ylabel('Conductivity (S/cm)')
plt.title('Mean Process Results')

```
