```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

grain_file = np.genfromtxt("conductivity data.csv", delimiter=",")

process1 = grain_file[2:42, 0], grain_file[2:42, 1] # Separates CSV into one array per process, taking the columns corresponding to that processes grain size and conductivity data
process2 = grain_file[2:42, 2], grain_file[2:42, 3]
process3 = grain_file[2:42, 4], grain_file[2:42, 5]
process4 = grain_file[2:42, 6], grain_file[2:42, 7]
process5 = grain_file[2:42, 8], grain_file[2:42, 9]


# Scatter plots
plt.figure (figsize=(15,10))

plt.subplot(231)    # Creates a subplot with two rows and three columns and sets our current position to the top left-plot
plt.scatter(process1[0], process1[1], s = 100,  color = 'red', marker = 'o', ec = 'black')  # plots a scatter graph of grain size as the x-axis, conductivity as the y-axis, sets the marker size, shape, colour, and outline
plt.title('Process 1')
plt.ylabel('Conductivity (S/m)')
plt.xlabel('Grain Size (um)')

plt.subplot(232)
plt.scatter(process2[0], process2[1], s = 100,  color = 'green', marker = '^', ec = 'black')
plt.title('Process 2')
plt.ylabel('Conductivity (S/m)')
plt.xlabel('Grain Size (um)')

plt.subplot(233)
plt.scatter(process3[0], process3[1], s = 100,  color = 'blue', marker = 'd', ec = 'black')
plt.title('Process 3')
plt.ylabel('Conductivity (S/m)')
plt.xlabel('Grain Size (um)')

plt.subplot(234)
plt.scatter(process4[0], process4[1], s = 100,  color = 'yellow', marker = 'p', ec = 'black')
plt.title('Process 4')
plt.ylabel('Conductivity (S/m)')
plt.xlabel('Grain Size (um)')

plt.subplot(235)
plt.scatter(process5[0], process5[1], s = 100,  color = 'purple', marker = 's', ec = 'black')
plt.title('Process 5')
plt.ylabel('Conductivity (S/m)')
plt.xlabel('Grain Size (um)')
```

```python
mean_grain = []
std_grain = []
# Creates mean and standard deviation for grain size in each process
for i in range (0,10,2): # iterates through columns two at a time (even columns are grain size, odd columns are conductivity)
    mean_grain = np.append(mean_grain,np.mean(grain_file[2:42,i]))  # uses numpy mean function to calculate mean for grain size of a particular process (all relevant rows of the current column we've iterated to)
    std_grain = np.append(std_grain,np.std(grain_file[2:42,i]))


mean_cond = []
std_cond = []
# Creates mean and standard deviation for grain conductivity in each process
for i in range(1,10,2):
    mean_cond = np.append(mean_cond,np.mean(grain_file[2:42,i]))
    std_cond = np.append(std_cond, np.std(grain_file[2:42,i]))
    
plt.figure (figsize=(15,10))
plt.errorbar(mean_grain, mean_cond, yerr = std_cond, xerr = std_grain, fmt = 'ro-', linewidth = 2, capsize = 6) # plots a line graph with error bars, using our mean as the points and the standard deviation in both grain size and conductivity to give us error bars in y and x directions
# fmt formats the markers and lines, r means red, o is the marker style, and - specifies connecting points with a solid line, line width controls the width of the connecting line and capsize defines the size of the caps on the error bars
plt.xlabel('Grain Size (um)')
plt.ylabel('Conductivity (S/cm)')
plt.title('Mean Process Results')
```
